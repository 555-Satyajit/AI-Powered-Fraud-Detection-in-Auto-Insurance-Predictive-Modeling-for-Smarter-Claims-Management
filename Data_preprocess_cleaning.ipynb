```python
import pandas as pd
import numpy as np

# Step 1: Print shape and preview
print("📊 Initial shape:", merged_df.shape)
print("🔍 Missing values before cleaning:\n", merged_df.isnull().sum()[merged_df.isnull().sum() > 0])

# Step 2: Identify column types
numeric_cols = merged_df.select_dtypes(include=[np.number]).columns
categorical_cols = merged_df.select_dtypes(include=['object', 'category', 'bool']).columns
date_cols = [col for col in merged_df.columns if 'date' in col.lower() or 'Date' in col]

# Step 3: Fill numeric columns with mean
for col in numeric_cols:
    if merged_df[col].isnull().any():
        merged_df[col] = merged_df[col].fillna(merged_df[col].mean())

# Step 4: Fill categorical columns with mode
for col in categorical_cols:
    if merged_df[col].isnull().any():
        mode_val = merged_df[col].mode()[0] if not merged_df[col].mode().empty else "Unknown"
        merged_df[col] = merged_df[col].fillna(mode_val)

# Step 5: Convert date columns to datetime format
for col in date_cols:
    merged_df[col] = pd.to_datetime(merged_df[col], errors='coerce')

# Step 6: Check final status
print("\n✅ Cleaning complete.")
print("🔄 Shape after cleaning:", merged_df.shape)
print("❓ Missing values remaining:\n", merged_df.isnull().sum()[merged_df.isnull().sum() > 0])
```

**5 IMPORTANT FEATURE ADDED**


```python
import numpy as np
import pandas as pd

# ✅ Ensure column names are standardized
merged_df.columns = merged_df.columns.str.strip().str.lower().str.replace(' ', '_')

# ✅ Convert necessary date columns
merged_df['claims_date'] = pd.to_datetime(merged_df['claims_date'], errors='coerce')
merged_df['accident_date'] = pd.to_datetime(merged_df['accident_date'], errors='coerce')
merged_df['policy_start_date'] = pd.to_datetime(merged_df['policy_start_date'], errors='coerce')

# ✅ Feature 1: Days between accident and claim
merged_df['days_accident_to_claim'] = (merged_df['claims_date'] - merged_df['accident_date']).dt.days

# ✅ Feature 2: Policy age at time of accident
merged_df['policy_age_at_accident_days'] = (merged_df['accident_date'] - merged_df['policy_start_date']).dt.days

# ✅ Feature 3: Total Claim to Vehicle Cost Ratio
merged_df['total_claim_to_vehicle_cost_ratio'] = (
    merged_df['total_claim'] / merged_df['vehicle_cost'].replace(0, np.nan)
).fillna(0)

# ✅ Feature 4: Number of Claim Types (Injury, Property, Vehicle)
if 'property_damage' in merged_df.columns:
    merged_df['property_damage_numeric'] = merged_df['property_damage'].apply(
        lambda x: 1 if str(x).strip().upper() == 'YES' else 0)
else:
    merged_df['property_damage_numeric'] = 0  # fallback

merged_df['num_claim_types'] = (
    (merged_df['injury_claim'] > 0).astype(int) +
    (merged_df['property_damage_numeric'] > 0).astype(int) +
    (merged_df['vehicle_claim'] > 0).astype(int)
)
merged_df.drop(columns=['property_damage_numeric'], inplace=True)

# ✅ Feature 5: Claim Delay Flag (>7 days)
merged_df['claim_delay_flag'] = (merged_df['days_accident_to_claim'] > 7).astype(int)

# ✅ Done: Show the new features
print("✅ Most important 5 features created:")
print(merged_df[[
    'days_accident_to_claim',
    'policy_age_at_accident_days',
    'total_claim_to_vehicle_cost_ratio',
    'num_claim_types',
    'claim_delay_flag'
]].head())
```

```python
import pandas as pd
from datetime import datetime

# Sample dummy data
data = {
    'Claim_Status': ['Approved', 'Rejected', 'Approved', 'Approved'],
    'Claims_Date': [datetime(2023, 5, 1), datetime(2023, 5, 3), datetime(2023, 5, 5), datetime(2023, 5, 7)],
    'Accident_Date': [datetime(2023, 4, 28), datetime(2023, 4, 30), datetime(2023, 5, 2), datetime(2023, 5, 5)],
    'Is_Fraud': [0, 1, 0, 1],
    'Claim_Amount': [5000, 2000, 3000, 4000],
    'Policy_ID': ['P1', 'P2', 'P3', 'P1'],
    'Feedback': ['Positive', 'Negative', 'Positive', 'Positive']
}

df_processed = pd.DataFrame(data)

# KPI 1: Claim Approval Rate
claim_approval_rate = (df_processed['Claim_Status'] == 'Approved').sum() / len(df_processed) * 100

# KPI 2: Average Claim Processing Time (in days)
df_processed['Processing_Time_Days'] = (df_processed['Claims_Date'] - df_processed['Accident_Date']).dt.days
avg_processing_time = df_processed['Processing_Time_Days'].mean()

# KPI 3: Fraud Detection Rate
fraud_detection_rate = df_processed['Is_Fraud'].sum() / len(df_processed) * 100

# KPI 4: Claim Cost per Policy
total_claim_cost = df_processed['Claim_Amount'].sum()
unique_policies = df_processed['Policy_ID'].nunique()
claim_cost_per_policy = total_claim_cost / unique_policies if unique_policies > 0 else 0

# KPI 5: Customer Satisfaction Score (CSAT)
csat_score = (df_processed['Feedback'] == 'Positive').sum() / df_processed['Feedback'].count() * 100

# Print results
print("Claim Approval Rate:", claim_approval_rate)
print("Avg Claim Processing Time (days):", avg_processing_time)
print("Fraud Detection Rate:", fraud_detection_rate)
print("Claim Cost per Policy:", claim_cost_per_policy)
print("Customer Satisfaction Score (CSAT):", csat_score)
```
